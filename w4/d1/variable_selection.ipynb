# Variable Selection

```python
import pandas as pd
import numpy as np
```

```python
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = "all"
```

```python
df_numeric = pd.read_csv("df_numeric.csv")
df_numeric.head()
```

### drop the target variable from the dataframe before doing feature selection so we don't change it accidentally

```python
target = df_numeric["SalePrice"]
df_numeric.drop("SalePrice", axis=1, inplace=True)
```

### Removing features with small variance

small variance = small predicting power

```python
from sklearn.feature_selection import VarianceThreshold
```

```python
vt = VarianceThreshold(0.1)
df_transformed = vt.fit_transform(df_numeric)
```

```python
df_numeric.shape
df_transformed.shape

# 9 features were removed
```

```python
df_numeric.head(2)
df_transformed
```

**Turning df_transformed back into a dataframe**

```python
vt.get_support()

df_numeric.columns

df_numeric.columns[vt.get_support()]
```

```python
df_transformed = pd.DataFrame(df_transformed, columns=df_numeric.columns[vt.get_support()])
df_transformed.head()
```

### Removing Correlated (Redundant) Features

```python
df_corr = df_transformed.corr().abs()
df_corr.head()
```

```python
indices = np.where(df_corr > 0.8)
indices
```

```python
type(indices)
```

```python
# These for loops produce the same result

for x, y in zip(indices[0], indices[1]):
    print(x, y)
    
for x, y in zip(*indices):
    print(x, y)
```

```python
index_pairs = []

for x, y in zip(*indices):
    if (x != y) and (x < y):
        index_pairs.append((df_corr.index[x], df_corr.columns[y]))
        
index_pairs
```

```python
for idx in index_pairs:
    try:
        df_transformed.drop(idx[1], axis=1, inplace=True)
    except KeyError:
        pass
```

```python
df_transformed.shape

# 5 more features dropped
```

### Forward Regression. Selecting the K best features (K features most correlated with the target variable)

```python
from sklearn.feature_selection import f_regression, SelectKBest
```

```python
skb = SelectKBest(f_regression, k=10)
X = skb.fit_transform(df_transformed, target)
X
```

```python
skb.get_support().shape

df_transformed.shape
```

```python
X = pd.DataFrame(X, columns=df_transformed.columns[skb.get_support()])
X
```

# Linear Regression with statsmodels

```python
import statsmodels.api as sm
```

```python
X = sm.add_constant(X)
```

```python
X
```

```python
y = target
```

```python
lin_reg = sm.OLS(y, X)
```

```python
type(lin_reg)
```

```python
model = lin_reg.fit()
```
