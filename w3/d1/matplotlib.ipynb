### Creating line plot example - Stateful

```python
x = np.linspace(0, 2, 100)

fig2 = pyplot.figure()

pyplot.plot(x, label="Linear")
pyplot.plot(x**2, label="Quadratic")
pyplot.plot(x**3, label="Cubic")

pyplot.xlabel("domain")
pyplot.ylabel("range")

pyplot.title("Linear, Quadratic, and Cubic Lines")

pyplot.legend()

pyplot.show()
```

# Basic Graphs

## Line Plot

```python
x = np.arange(0,100,3.14159/10)
y = np.sin(x)

fig = pyplot.figure(figsize=(10,6))

pyplot.plot(x, y, color="red")

pyplot.xticks(np.arange(0,101,10))

pyplot.show()
```

```python
x = np.random.rand(200,1)
y = np.random.rand(200,1)

fig2 = pyplot.figure(figsize=(10,6))

pyplot.scatter(x,y)

pyplot.show()
```

```python
y = 15*np.random.randn(200,1)

fig3 = pyplot.figure(figsize=(10,6))

pyplot.hist(y, bins=20)

pyplot.show()
```

# Customizations

## Plot 2 lines on same figure

```python
x = np.linspace(0,100,200)
y1 = np.sin(x)
y2 = np.cos(x)

fig4 = pyplot.figure(figsize=(10,6))

pyplot.plot(x, y1, color="red")
pyplot.plot(x, y2, color="blue")

pyplot.show()
```

## Plot 2 lines on separate figures using subplots

```python
x = np.linspace(0,100,200)
y1 = np.sin(x)
y2 = np.cos(x)

fig5 = pyplot.figure(figsize=(10,6))

pyplot.subplot(2,1,1)
pyplot.plot(x, y1, color="r")
pyplot.title("sin(x)")

pyplot.subplot(2,1,2)
pyplot.plot(x, y2, color="b")
pyplot.title("cos(x)")

pyplot.tight_layout() # I think this fits the figure to the screen? No it does not but it changes the size somehow

pyplot.show()
```

## Legends

```python
# import pandas
import pandas as pd

# import sklearn datasets
from sklearn import datasets
```

```python
iris = datasets.load_iris()
iris_df = pd.DataFrame(iris.data, columns=iris.feature_names)
iris_df["target"] = iris.target
iris_df["target_name"] = iris_df.target.map({0:"setosa", 1:"versicolor", 2:"virginica"})
```

```python
iris_df.head()
```

Making a scatter plot comparing sepal length to width for each type of flower

```python
setosa_df = iris_df[iris_df["target_name"].eq("setosa")]
versicolor_df = iris_df[iris_df["target_name"].eq("versicolor")]
virginica_df = iris_df[iris_df["target_name"].eq("virginica")]
```

```python
pyplot.scatter(x=setosa_df["sepal length (cm)"], y=setosa_df["sepal width (cm)"],
              color="red", label="setosa")
pyplot.scatter(x=versicolor_df["sepal length (cm)"], y=versicolor_df["sepal width (cm)"],
              color="green", label="versicolor")
pyplot.scatter(x=virginica_df["sepal length (cm)"], y=virginica_df["sepal width (cm)"],
              color="blue", label="virginica")

pyplot.legend(loc="lower right")
pyplot.title("Iris Flower")
pyplot.ylabel("Sepal Width")
pyplot.xlabel("Sepal Height")

pyplot.show()
```

### Adding Annotations

all the same code as before except `pyplot.annotate()` lines added

```python
pyplot.scatter(x=setosa_df["sepal length (cm)"], y=setosa_df["sepal width (cm)"],
              color="red", label="setosa")
pyplot.scatter(x=versicolor_df["sepal length (cm)"], y=versicolor_df["sepal width (cm)"],
              color="green", label="versicolor")
pyplot.scatter(x=virginica_df["sepal length (cm)"], y=virginica_df["sepal width (cm)"],
              color="blue", label="virginica")

pyplot.legend(loc="lower right")
pyplot.title("Iris Flower")
pyplot.ylabel("Sepal Width")
pyplot.xlabel("Sepal Height")

pyplot.annotate("Setosa", xy=(5.0,3.5), xytext=(4.25,4.0), arrowprops={})
pyplot.annotate("Versicolor", xy=(5.97,2.65), xytext=(5.2,2.1), arrowprops={})
pyplot.annotate("Virginica", xy=(6.4,3.4), xytext=(6.9,4.0), arrowprops={"color":"red"})

pyplot.show()
```
