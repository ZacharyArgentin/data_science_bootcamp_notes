# SEABORN!

```python
import numpy as np
import pandas as pd
from matplotlib import pyplot
import seaborn as sns
```

```python
help(sns.set_theme)
```

```python
sns.get_dataset_names()
```

```python
df = sns.load_dataset("flights")
df.head()
```

### relplot()

used for scatter and line plots

```python
sns.relplot(x="month", y="passengers", data=df, kind="line")
```

```python
sns.relplot(x="month", y="passengers", data=df, kind="scatter", hue="year")
```

```python
sns.set_theme(style="darkgrid")
sns.relplot(x="month", y="passengers", hue="year", palette="ch:r=-.5,l=.75", data=df)
```

### catplot()

```python
df_tips = sns.load_dataset("tips")
```

```python
df_tips.head()
```

```python
sns.catplot(x="day", y="total_bill", data=df_tips, kind="box")
```

### facet_grid()

```python
iris = sns.load_dataset("iris")
iris.head()
```

# WALKTHRU

```python
df_tips.head()
```

`lmplot()` plots a regression line with a confidence interval

```python
sns.lmplot(x='total_bill', y='tip', data=df_tips)
pyplot.show()
```

```python
sns.lmplot(x="total_bill", y="tip", data=df_tips, hue="sex")
pyplot.show()
```

```python
sns.lmplot(x="total_bill", y="tip", data=df_tips, col="sex")
pyplot.show()
```

### Residual Plot

A residual plot shows by how much each data point missed the regression line

```python
sns.residplot(x="total_bill", y="tip", data=df_tips)
pyplot.show()
```

### Strip Plot

```python
sns.stripplot(x="day", y="tip", data=df_tips, size=5, jitter=True)

pyplot.ylabel("tip $")
pyplot.show()
```

### Swarm Plot

```python
pyplot.figure(figsize=(15,9))
sns.swarmplot(x='day',y='tip',data=df_tips, hue='sex')
pyplot.ylabel('tip ($)')
pyplot.show()
```

### Boxplot

```python
sns.boxplot(x="day", y="tip", data=df_tips)
pyplot.show()
```

# Joint Plot

```python
sns.jointplot(x="total_bill", y="tip", data=df_tips)
pyplot.show()
```

```python
sns.jointplot(x="total_bill", y="tip", data=df_tips, kind="kde");
```

### Pair plot

```python
sns.pairplot(df_tips);
```

```python
sns.pairplot(df_tips, hue="sex");
```
