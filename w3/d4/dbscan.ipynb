# DBSCAN - Density Based Clustering

```python
from sklearn.datasets import make_blobs
from matplotlib import pyplot
import numpy as np

from sklearn.datasets import make_moons
from sklearn.cluster import DBSCAN

pyplot.rcParams["figure.figsize"]=(15,9)
```

```python
X, y = make_blobs(n_samples=150, n_features=2, centers=3, cluster_std=0.5, random_state=0)
```

```python
def plot_clusters(X, cluster_prediction, plot_cluster_centers=False):
    for cluster in set(cluster_prediction):
        x = X[(cluster_prediction == cluster), 0]
        y = X[(cluster_prediction == cluster), 1]
        pyplot.scatter(x, y, label=f"cluster {cluster}")
        
    if plot_cluster_centers:
        x_centroids = []
        y_centroids = []
        x_centroids.append(np.mean(x))
        y_centroids.append(np.mean(y))
        pyplot.scatter(x_centroids, y_centroids, c="red", marker="*")
        
    pyplot.legend()
    pyplot.grid()
    pyplot.show()
```

```python
db = DBSCAN(eps=0.5, min_samples=5, metric="euclidean")

y_db = db.fit_predict(X)
```

```python
plot_clusters(X, y_db)
```

### Moon shaped data

```python
X, y = make_moons(n_samples=200, noise=0.05, random_state=0)
```

```python
pyplot.scatter(X[:,0], X[:,1]);
```

### Clustering using KMeans

```python
from sklearn.cluster import KMeans
```

```python
km = KMeans(n_clusters=2, random_state=0)
y_km = km.fit_predict(X)
```

```python
plot_clusters(X, y_km)
```

### Clustering using DBSCAN

```python
db = DBSCAN(eps=0.2)
y_db = db.fit_predict(X)
```

```python
plot_clusters(X, y_db)
```
