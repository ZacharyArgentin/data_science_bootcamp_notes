## Density Models Assignment

```python
# import matplotlib
from matplotlib import pyplot

# set figure size
pyplot.rcParams["figure.figsize"] = (15,9)

#import numpy
import numpy as np

# from sklearn import make_blobs
from sklearn.datasets import make_blobs

# import DBSCAN
from sklearn.cluster import DBSCAN
```

## Task: Generate clusters with make_blobs function 

Use the following settings:
   * n_samples=150,
   * n_features=2,
   * centers=3,
   * cluster_std=0.6,
   * random_state=0

```python
X, y = make_blobs(n_samples=150, n_features=2, centers=3, cluster_std=0.6, random_state=0)
```

* plot data points

```python
pyplot.scatter(X[:,0], X[:,1]);
```

* define function plot_clusters

```python
def plot_clusters(X, predict, params):
    for cluster in set(predict):
        x = X[predict == cluster, 0]
        y = X[predict == cluster, 1]
        pyplot.scatter(x, y, label=f"cluster {cluster}")
    
    pyplot.title(f"eps={params[0]}, min_samples={params[1]}")
    pyplot.legend()
    pyplot.show()
```

## Objective: 
Try to find optimal settings of `eps` and `min_samples` params of DBSCAN. You can iterate over multiple parameters combinations and plot the clustering results.

* define params space

```python
params=(0.5, 5)
```

```python
db = DBSCAN(eps=params[0], min_samples=params[1])
cluster_membership = db.fit_predict(X)
```

```python
plot_clusters(X, cluster_membership, params)
```

* combinations (hint: use 'product' function from 'itertools')

```python
from itertools import product
```

```python
param_combinations = product(np.round(np.arange(0.4, 0.7, 0.1), 2), range(5, 9, 1))
```

```python
for combination in param_combinations:
    db = DBSCAN(eps=combination[0], min_samples=combination[1])
    cluster_membership = db.fit_predict(X)
    plot_clusters(X, cluster_membership, combination)
```
