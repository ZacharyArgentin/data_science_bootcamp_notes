```python
fig, ax = pyplot.subplots(figsize=(10,6))

fig.patch.set_facecolor("lightsteelblue")

ax.plot(x, y, c="red")
ax.set_title("Title", fontdict={"fontsize": 24, "c":"snow", "alpha":0.6})
ax.grid()
ax.set_xticks(range(5,101,5))

pyplot.show()
```
