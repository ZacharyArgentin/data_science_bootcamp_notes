# Linear Transformations & Eigenthings

```python
import numpy as np
from matplotlib import pyplot

pyplot.style.use("fivethirtyeight")
```

Matrices serve as linear trasformations

```python
# Create transformation matrix
A = np.array([[-1,3], [2, -2]])
```

```python
# Create vector
v = np.array([[2], [1]])
```

```python
def plotVectors(vecs, colors, alpha=1, figsize=(10,8)):
    """
    Plot set of vectors.

    Parameters
    ----------
    vecs : array-like
        Coordinates of the vectors to plot. Each vectors is in an array. For
        instance: [[1, 3], [2, 2]] can be used to plot 2 vectors.
    colors : array-like
        Colors of the vectors. For instance: ['red', 'blue'] will display the
        first vector in red and the second in blue.
    alpha : float
        Opacity of vectors
    figsize : tuple (x_size,y_size)
        Size of the figure

    Returns:

    fig : instance of matplotlib.figure.Figure
        The figure of the vectors
    """
    
    pyplot.figure(figsize=figsize)
    pyplot.axvline(x=0, color="grey", zorder=0)
    pyplot.axhline(y=0, color="grey", zorder=0)
    
    vecs = [vec.flatten() for vec in vecs]
    
    for i in range(len(vecs)):
        x = np.concatenate([[0,0], vecs[i]])
        pyplot.quiver([x[0]], [x[1]], [x[2]], [x[3]],
                      angles="xy", scale_units="xy", scale=1, color= colors[i], alpha=alpha)
```

```python
A
```

```python
v
```

```python
Av = A.dot(v)
Av
```

```python
plotVectors([v, Av], colors=["lightblue", "yellow"])
pyplot.ylim(-1,4)
pyplot.xlim(-1,4)
pyplot.show()
```

# Eigenvector

A transformation matrix that changes the size of the vector but not the direciton

```python
A = np.array([[5,1], [3,3]])
A
```

```python
# eigenvector
v1 = np.array([[1],[1]])
v1
```

```python
Av1 = A.dot(v1)
Av1
```

```python
plotVectors([Av1, v1], colors=["lightblue", "yellow"])
pyplot.show()
```

```python
v2 = np.array([[1], [-3]])
v2
```

```python
Av2 = A.dot(v2)
Av2
```

```python
plotVectors([Av2, v2], colors=["blue", "yellow"])
pyplot.show()
```

# Finding Eigenvalues with Numpy

> Numpy linalg module provides a function `eig()` returning eigenvectors and eigenvalues.

```python
A
```

```python
eigenvals, V = np.linalg.eig(A)
eigenvals
```

```python
V
```

```python
lambdas = np.diag(eigenvals)
lambdas
```

```python
V_inv = np.linalg.inv(V)
V_inv
```

```python
A_orig  = V.dot(lambdas).dot(V_inv)
A_orig
```
