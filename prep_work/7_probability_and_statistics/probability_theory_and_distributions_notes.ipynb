### Binomial Distribution

```python
# n = no. of trials
# p = probability of successful outcome
# size = no. of repeats


n = 10
p = 0.5
size = 1000

samples = np.random.binomial(n, p, size)
print(samples)

# The result shows the number of succesful outcomes after n amount of trials repeated 'size' amount of times
```

```python
# Counter looks at the samples array and counts the number of times each value occured, and then puts them in a dictionary
head_occu_count = Counter(samples)
print(head_occu_count)

# .values() yields a list of the values in the dicitonary
print(head_occu_count.values())

# Make a list of the probability of getting heads x amount of times by taking the amount of times it actually happened
# divided by the total number of trials.
probability_of_getting_heads = [value/size for value in head_occu_count.values()]
print(probability_of_getting_heads)
```

graph the results

```python
plt.scatter(x=head_occu_count.keys(), y=probability_of_getting_heads)
plt.xlabel("Number of heads out of 10 tosses")
plt.ylabel("Probability")
plt.show()
```

### Normal Distribution

```python
samples = np.random.normal(loc=0, scale=1, size=1000000)
print(samples)
```

```python
plt.hist(samples)
plt.show()

plt.hist(samples, density=True, histtype="step")
plt.show()

plt.hist(samples, density=True, histtype="step")
plt.show()

plt.hist(samples, density=True, histtype="step", bins=100)
plt.show()
```

### Changing the standard deviation and the mean

changing the standard deviation

```python
samples_std1  = np.random.normal(loc=0, scale=1, size=1000000)
samples_std3  = np.random.normal(loc=0, scale=3, size=1000000)
samples_std10 = np.random.normal(loc=0, scale=10, size=1000000)

plt.hist(samples_std1, density=True, histtype="step", bins=100)
plt.hist(samples_std3, density=True, histtype="step", bins=100)
plt.hist(samples_std10, density=True, histtype="step", bins=100)

# make a legend, set limits, and show the plot
plt.legend(('std = 1', 'std = 3', 'std = 10'))
plt.ylim(-0.01, 0.42)
plt.show()
```

changing the mean

```python
samples_std1  = np.random.normal(loc=10, scale=1, size=1000000)
samples_std3  = np.random.normal(loc=20, scale=3, size=1000000)
samples_std10 = np.random.normal(loc=30, scale=10, size=1000000)

plt.hist(samples_std1, density=True, histtype="step", bins=100)
plt.hist(samples_std3, density=True, histtype="step", bins=100)
plt.hist(samples_std10, density=True, histtype="step", bins=100)

# make a legend, set limits, and show the plot
plt.legend(('std = 1', 'std = 3', 'std = 10'))
plt.ylim(-0.01, 0.42)
plt.show()
```
