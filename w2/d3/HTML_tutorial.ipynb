### HTML Tutorial

```python
import pandas as pd
import requests
```

```python
url = 'https://www.worldcoinindex.com/'
crypto_url = requests.get(url)
crypto_url
```

**Save full HTML source code of our webpage into variable body** (using `.text' attribute of response object)

```python
html_source_code = crypto_url.text
```

```python
html_source_code
```

### Pandas can extract a dataframe from html code if the code has \<table> tags

```python
crypto_data = pd.read_html(html_source_code)
print(type(crypto_data))
print(len(crypto_data))
crypto_data
```

```python
crypto_data = crypto_data[0]
crypto_data.head()
```
