```python
import numpy as np
import pandas as pd
```

```python
dates = pd.date_range('20210629', periods=6)
df = pd.DataFrame(np.random.randn(6, 4), index=dates, columns=list('ABCD'))
df2 = pd.DataFrame({'A': 1.,
                        'B': pd.Timestamp('20130102'),
                        'C': pd.Series(1, index=list(range(4)), dtype='float32'),
                        'D': np.array([3] * 4, dtype='int32'),
                        'E': pd.Categorical(["test", "train", "test", "train"]),
                        'F': 'foo'})
df
```

### Selecting

**Selecting Column by Name**

```python
df["A"]
```

**Selecting Column by index**

```python
df.iloc[:,0]
```

**Selecting Row**

```python
df.loc["20210629"]
```

**Selecting Row by index**


```python
df.iloc[0]
```

### Slicing

**Slice Rows**

```python
df["20210701" : "20210703"]
```

```python
df.iloc[2:5]
```

**Slice Columns**

```python
df.loc[:,"A":"C"]
```

```python
df.iloc[:,0:3]
```

**Slice Rows & Columns**

```python
df.loc["2021-06-30":"2021-07-02", "B":"D"]
```

You can also substitute the slices for lists to get rows/columns that are not sequential

### Select by dtypes

```python
df = pd.DataFrame({'string': list('abc'),
                       'int64': list(range(1, 4)),
                       'uint8': np.arange(3, 6).astype('u1'),
                       'float64': np.arange(4.0, 7.0),
                       'bool1': [True, False, True],
                       'bool2': [False, True, False],
                       'dates': pd.date_range('now', periods=3),
                       'category': pd.Series(list("ABC")).astype('category')})
df
```

```python
df.select_dtypes(["bool"])
```

```python
df.select_dtypes(["int","float"])
```

### Boolen indexing (i.e. filter arrays)

```python
df
```

```python
df[df['float64'] >= 5]
```

```python
df2 = df.copy()
df2['E'] = ['one', 'two', 'three']
df2
```

```python
df2

df2["E"].isin(["one", "three"])

df2[df2["E"].isin(["one", "three"])]
```

### Setting Values

```python
df3 = df.copy()
```

```python
df
```

```python
df3.iat[0, 1] = -1
df3

df3.iloc[0, 1] = -2
df3
```

```python
df3.at[0, 'float64'] = -10
df3
df3.loc[0, 'float64'] = -20
df3
```

```python
df3.loc[:, 'uint8'] = np.array([50] * len(df))
df3
```
