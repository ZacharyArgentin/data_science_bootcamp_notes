You can find them [here](https://drive.google.com/file/d/1NY6cmF9Shjw-dD7BD6bNmfcIVz-kQcFR/view?usp=sharing).

```python
cast = pd.read_csv('data/cast.csv', index_col=None)
cast.head()
```

```python
release_dates = pd.read_csv('data/release_dates.csv', index_col=None,
                                      parse_dates=['date'], infer_datetime_format=True)
release_dates.head()
```

```python
day_num_to_day_name = {0:"Monday", 1:"Tuesday", 2:"Wednesday", 3:"Thursday", 4:"Friday", 5:"Saturday", 6:"Sunday"}
```

```python
month_num_to_month_name = {1:"January", 2:"February", 3:"March", 4:"April", 5:"May", 6:"June",
                           7:"July", 8:"August", 9:"September", 10:"October", 11:"November", 12:"December"}
```

### Count the number of movies with "Christmas" in their title for each month, that is released are the USA.

```python
release_dates[release_dates["title"].str.contains("Christmas") & release_dates["country"].eq("USA")].groupby(release_dates["date"].dt.month).count().sort_values("title", ascending=False)["title"].rename_axis(index="month").rename("# of movies")
```

### Count the number of movies with "Christmas" in their title for each month, that are released in Canada.

```python
release_dates["month"] = release_dates["date"].dt.month_name()
release_dates.head()
```

```python
release_dates[release_dates["title"].str.contains("Christmas") & release_dates["country"].eq("Canada")].value_counts("month").sort_index()
```

### Count the number of movies in which the titles start with "The Hobbit" for each month, that are released in the USA.

```python
release_dates[release_dates["title"].str.startswith("The Hobbit") &
             release_dates["country"].eq("USA")].groupby("month").size()
```

### Count the number of movies with "Romance" in their title for **each day of the week**, that are released in the USA.

```python
release_dates[release_dates["title"].str.contains("Romance") & release_dates["country"].eq("USA")]\
.value_counts(release_dates["date"].dt.day_name())
```

### Count the number of movies with "Action" in their title for **each day of the week**, that are released in the USA.

```python
action_movies = release_dates[release_dates["title"].str.contains("Action") & release_dates["country"].eq("USA")]
action_movies.head()
```

```python
action_movies.value_counts(action_movies["date"].dt.day_name())
```

```python
action_count = action_movies.value_counts(action_movies["date"].dt.weekday)
action_count.sort_index().rename(index=day_num_to_day_name)
```

### In which months do films with the actress Judi Dench tend to be released in the USA?

```python
pd.merge(cast[cast["name"].eq("Judi Dench")],release_dates[release_dates["country"].eq("USA")])
```
